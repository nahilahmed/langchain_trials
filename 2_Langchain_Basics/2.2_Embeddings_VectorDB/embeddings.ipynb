# Embedding Techniques

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(
    model="text-embedding-3-large"
)

embeddings
```

```python
text = "This is a test sentence for embedding."
embedding = embeddings.embed_query(text)
print(f"Embedding for the text '{text}': {embedding}")

# get the dimensions of the embedding
print(f"Embedding dimensions: {len(embedding)}")
```

```python
# What if I want to reduce the dimensions of the embedding?

embeddings_1024 = OpenAIEmbeddings(
    model="text-embedding-3-large",
    dimensions=1024
)

embedding_1024_vector = embeddings_1024.embed_query(text)
# get the dimensions of the reduced embedding
print(f"Reduced embedding dimensions: {len(embedding_1024_vector)}")
```
